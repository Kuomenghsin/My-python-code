新增(insert)：

每一次新增(insert)都會呼叫class TreeNode，賦予節點三種屬性(val、left、right)

主要分為以下幾種case：

1.如果root為空(前面可能有先delete root)

2.如果新增的值與"某節點"的val重複，加在"某節點"的left

3.如果新增的值比root大，往右邊進行recursive

4.如果新增的值比root小，往左邊進行recursive

5.最後return 新增的值的節點(output會看到<__main__.TreeNode at 0x2c75082b808>)

6.由於節點本身就是一個記憶體的位置，裡面包含三種屬性(val、left、right)，所以節點 "不等於" 節點的值

查詢(search)：

1.與新增非常類似，都要先進行走訪，只是查詢到最後是直接return target節點

2.如果查詢的值跟root一樣，直接return root，不必管有沒有重複值的case

3.如果查詢的值比root大，往右邊進行recursive

4.如果查詢的值比root小，往左邊進行recursive

5.最後return target值節點(output會看到<__main__.TreeNode at 0x2c75082b808>)

6.由於節點本身就是一個記憶體的位置，裡面包含三種屬性(val、left、right)，所以節點 "不等於" 節點的值

刪除(delete)：


這是邏輯架構最複雜的部份，需要考慮的case有十幾種，依據每個人的分類粗細不同

主要分為以下幾種case：

1.刪除值在root節點：
    
    1-1：若有重複值，先把重複值弄成只有一個
    
    1-2：root底下沒有小孩，return None
    
    1-3：root底下只有左小孩，把左小孩指定成root，並且return root
    
    1-4：root底下只有右小孩，把右小孩指定成root，並且return root
    
    1-5：root底下有兩個小孩
        
        1-5-1：root.left.right 跟 root.right.left 皆為空 找root.left補上來(或者root.right補上來也可)
        
        1-5-2：root.left.right 為空 且 root.right.left 存在，找root.right.left.left到底的節點來補刪除節點的位置
        
        1-5-3：root.left.right 存在(root.right.left存在也沒差)，找root.left.right.right到底的節點來補刪除節點的位置

2.刪除值不在root節點：
    
    2-1：先找到刪除值的節點
    
    2-2：再往左邊找有沒有重複值
    
    2-3：刪除值的節點底下只有左小孩
        
        2-3-1：左小孩底下 沒有右小孩 直接把刪除節點的parent跟left連結起來
        
        2-3-2：左小孩底下 有右小孩 找右小孩當中最大的來補刪除節點的位置
    
    2-4：刪除值的節點底下只有右小孩
        
        2-4-1：右小孩底下 沒有左小孩 直接把刪除節點的parent跟right連結起來
        
        2-4-2：右小孩底下 有左小孩 找左小孩當中最小的來補刪除節點的位置
    
    2-5：刪除值的節點底下有兩個小孩
        
        2-5-1：.left.right 跟 .right.left 皆為空 找.left補上來(或者.right補上來也可)
        
        2-5-2：.left.right 為空 且 .right.left 存在，找.right.left.left到底的節點來補刪除節點的位置
        
        2-5-3：.left.right 存在(root.right.left存在也沒差)，找.left.right.right到底的節點來補刪除節點的位置
    
    2-6：刪除值的節點底下沒有小孩，直接把節點指定為None      
        
修改(modify)

1.如果target == new_val，BST結構不變，直接return root(當然也可以先delete target再insert new_val，但時間複雜度高)

2.確定target != new_val，先insert new_val，再呼叫delete，並且把delete reuturn的結果return出來(新root)
