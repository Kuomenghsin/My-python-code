```python
from collections import defaultdict

class Graph(): 

    def __init__(self, vertices): 
        self.V = vertices 
        self.graph = [] 
        self.array = []
        self.roots = {}
    
    def addEdge(self,u,v,w): 
        if not self.array:
            self.array.append([u,v,w])
        else:
            det = True
            for i in range(len(self.array)):
                if self.array[i][2] > w:
                    self.array.insert(i,[u,v,w])
                    det = False
                    break
            if det:
                self.array.append([u,v,w])

    def K_helper(self,final):
        
        pivot = self.array[0]

        while self.roots[pivot[0]] != None and self.roots[pivot[1]] != None and self.roots[pivot[0]] == self.roots[pivot[1]]:
            self.array.pop(0)
            pivot = self.array[0]
        
        self.array.remove(pivot)
        
        small = min(pivot[0],pivot[1])
        large = max(pivot[0],pivot[1])

        if self.roots[large] == None and self.roots[small] == None:
            self.roots[large] , self.roots[small] = small , small

        elif self.roots[large] == None and self.roots[small] != None:
            self.roots[large] = self.roots[small]          

        else:
            for i in range(self.V):
                if i != large and self.roots[i] == self.roots[large]:
                    self.roots[i] = small
            self.roots[large] , self.roots[small] = small , small

        final[str(str(pivot[0])+"-"+str(pivot[1]))] = pivot[2]

        if len(final) != self.V-1:
            self.K_helper(final)
        
        return final
    
    def Kruskal(self):
        for j in range(self.V): 
            self.roots[j] = None
        final = {}
        return self.K_helper(final)
    
    def Dijkstra(self, s):
        visited = [s]
        final = dict()
        parent = dict()
        temp = []
        for i in range(self.V):
            final[i] = None
            parent[i] = None
        final[s] = 0
        for i in range(self.V-1):
            for j in range(len(self.graph[visited[-1]])):
                if self.graph[visited[-1]][j] != 0 and j not in visited:
                    
                    val = self.graph[visited[-1]][j] + final[visited[-1]]
                    
                    if not final[j]:
                        final[j] = val
                        parent[j] = visited[-1]
                    else:
                        if val < final[j]:
                            final[j] = val
                            parent[j] = visited[-1]
                    
                    if not temp:
                        temp.append([j,final[j]])
                    else:
                        det = True
                        for k in range(len(temp)):
                            if temp[k][1] > final[j]:
                                temp.insert(k,[j,final[j]])
                                det = False
                                break
                        if det:
                            temp.append([j,final[j]])
            
            while temp[0][0] in visited:
                temp.pop(0)
            visited.append(temp[0][0])
            temp.pop(0)
        
        last = {}
        for i in range(self.V):
            last[str(i)] = final[i]

        return last
```

Dijkstra、Kruskal原理說明：

"最短路徑" 跟 "最小生成樹" 的差別在於

最短路徑，尋找的答案是，給定一個基準點，要算出 "所有其他點" 到 "基準點" 的 "最短距離"

最小生成樹，不需要給定基準點，而是會從權重(或是路徑)最小的path開始走訪，直到連成V-1條邊

=================

最短路徑的執行步驟

1.從基準點開始，向外延伸一格能到達的所有vertex，並且從那些vertex當中，選一條路徑最短的

2.再從剛剛選定的vertex向外延伸一格所有能到達的vertex，並且也是選一條路徑最短的

3.重複2.這個步驟，直到每個vertex都被拜訪過，那麼最終的答案就出來了

4.拜訪過的vertex就不會再重複拜訪了

5.需要紀錄每個vertex的 "上一個節點" 或者是說parent，以便回頭尋找最短路徑所經過的點

程式碼實現的部分

1.創建final跟parent兩個deictionary，長度為self.V，一開始都預設為None

2.創建visited，用來存放 "拜訪過的節點" ，創建temp，其概念為 "Priority Queue"

3.承上，其概念為，按照最短路徑 "從小到大" 來排序，假設0到1長度為4，0到7長度為8，那麼temp = [[1,4],[7,8]]

4.依照Queue的概念，每次都只能從頭取出，如此變成確保取出的皆為最短路徑，還要注意取出的點不能在visited裡面

==================

最小生成樹的執行步驟

1.先把所有path按照權重，從小到大排序

2.需要一個表格來記錄每個edge所屬的 "subset" 或稱 "子集合" 至於子集合要用哪個edge當作代表 可以自行訂定一套規則

3.每次都會從權重最小的path選一個，並且確認那兩個edge的subset不相同，才可以把這個path加入最小生成樹

4.重複執行3.這個步驟，直到有V-1個path都被加入最小生成樹，那就可以停止了

程式碼實現的部分

1.由於要防止loop，會有discard的情況產生，無法確定程式碼執行的次數，故需要使用recursive的寫法，所以分2個def

2.創建self.array、self.roots(子集合)，使用self.變數，就可以在整個class內部使用，比較方便也不用去思考error

3.助教輸入的u v w，我會先弄成[u,v,w]，再按照w由小到大排，放入self.array，有點類似 "Priority Queue" 的概念

4.我自行定義的self.roots規則為，如果今天[7 6 1]要進來，7跟6都沒有root的情況，此兩個點的root都會變成6(選小的)

5.再來[6 5 2]要進來的話，7 6 5的root都會變成5，所以代表要重新去看 "有誰的root跟6的root一樣，請改成5哦"

6.創建final這個dictionary，預設長度為0，每次成功的新增path到最小升成樹之後，final都會新增一次

7.當final的長度還沒達到V-1的時候，就會一直進行recursive，反之，則可以return final

8.每次從self.array取出的時候，一樣是用Queue的概念，while取出的 "兩個邊的roots相同，就刪除，重新取出"
